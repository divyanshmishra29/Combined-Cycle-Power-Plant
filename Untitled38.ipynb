```
X=ccpp_data.drop(['PE'],axis=1)
```

```
y=ccpp_data['PE']
```

```
X.shape
```

```
y.shape
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.33, random_state=42)
```

```
from sklearn.linear_model import LinearRegression
```

```
m=LinearRegression()
```

```
m.fit(X_train,y_train)
```

```
pred=m.predict(X_test)
```

```
from sklearn.metrics import mean_squared_error
```

```
print(sqrt(mean_squared_error(y_test,pred)))
```

```
from sklearn.tree import DecisionTreeRegressor
```

```
m1=DecisionTreeRegressor()
```

```
m1.fit(X_train,y_train)
```

```
pred1=m1.predict(X_test)
```

```
print(sqrt(mean_squared_error(y_test,pred1)))
```

```
from sklearn.neighbors import KNeighborsRegressor
```

```
m2=KNeighborsRegressor()
```

```
m2.fit(X_train,y_train)
```

```
pred2=m2.predict(X_test)
```

```
print(sqrt(mean_squared_error(y_test,pred2)))
```

```
from sklearn.ensemble import RandomForestRegressor
```

```
m3=RandomForestRegressor()
```

```
m3.fit(X_train,y_train)
```

```
pred3=m3.predict(X_test)
```

```
print(sqrt(mean_squared_error(y_test,pred3)))
```

```
from sklearn.svm import SVR
```

```
m4=SVR()
```

```
m4.fit(X_train,y_train)
```

```
pred4=m4.predict(X_test)
```

```
print(sqrt(mean_squared_error(y_test,pred4)))
```

```
from sklearn.linear_model import Lasso
```

```
m5=Lasso()
```

```
m5.fit(X_train,y_train)
```

```
pred5=m5.predict(X_test)
```

```
print(sqrt(mean_squared_error(y_test,pred5)))
```

```
CCPP_eval = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/CCPP/Testing_set_ccpp.csv')
```
